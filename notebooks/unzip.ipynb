```python
import zipfile
from pathlib import Path

# Path setup
PROJECT_ROOT = Path("../")  # from notebooks → CTGA_01
DATASETS_DIR = PROJECT_ROOT / "datasets"

print("Datasets dir:", DATASETS_DIR.resolve())

zip_files = list(DATASETS_DIR.rglob("*.zip"))
print(f"Found {len(zip_files)} zip files\n")

for zip_path in zip_files:
    extract_to = zip_path.parent / zip_path.stem

    # skip if already extracted
    if extract_to.exists():
        print(f"Skipping (already extracted): {zip_path.name}")
        continue

    print(f"Extracting: {zip_path.name}")
    
    try:
        with zipfile.ZipFile(zip_path, 'r') as zip_ref:
            zip_ref.extractall(extract_to)
        print(f" → Extracted to: {extract_to}\n")
    except Exception as e:
        print(f"ERROR extracting {zip_path}: {e}")
```

```python
import os
from pathlib import Path

DATASETS = Path("../datasets").resolve()
IGNORE_FOLDERS = [".git", ".ipynb_checkpoints", "__MACOSX"]

def print_folder_structure(root: Path, prefix=""):
    """Recursively print folder structure ignoring files."""
    try:
        items = [d for d in root.iterdir() if d.is_dir() and d.name not in IGNORE_FOLDERS]
    except PermissionError:
        return
    except FileNotFoundError:
        return

    for i, folder in enumerate(sorted(items, key=lambda x: x.name.lower())):
        connector = "└── " if i == len(items) - 1 else "├── "
        print(prefix + connector + folder.name)
        # Recurse into the folder
        print_folder_structure(folder, prefix + ("    " if i == len(items) - 1 else "│   "))

# Print the folder structure for all datasets
print("DATASET STRUCTURE\n")
for dataset in sorted(DATASETS.iterdir(), key=lambda x: x.name.lower()):
    if dataset.is_dir():
        print(f"├── {dataset.name}")
        print_folder_structure(dataset, prefix="│   ")
```

```python
import zipfile
from pathlib import Path
import os

DATASETS_DIR = Path("../datasets")

zip_files = list(DATASETS_DIR.rglob("*.zip"))
print("Zips found:", len(zip_files))


def extract(zip_path, extract_to):
    try:
        with zipfile.ZipFile(zip_path, 'r') as z:
            z.extractall(extract_to)
        return
    except:
        pass

    # long path fallback
    zip_path_long = Path(r"\\?\{}".format(zip_path.resolve()))
    extract_to_long = Path(r"\\?\{}".format(extract_to.resolve()))

    with zipfile.ZipFile(zip_path_long, 'r') as z:
        for m in z.infolist():
            target = extract_to_long / m.filename
            if m.is_dir():
                os.makedirs(target, exist_ok=True)
            else:
                os.makedirs(target.parent, exist_ok=True)
                with open(target, "wb") as f:
                    f.write(z.read(m))


for z in zip_files:
    dest = z.parent / z.stem
    print("Extracting:", z.name)
    extract(z, dest)
    print("Done\n")
```

```python
import os

def print_folder_structure(root_path, indent=""):
    try:
        items = sorted(os.listdir(root_path))
    except PermissionError:
        return

    folders = [item for item in items if os.path.isdir(os.path.join(root_path, item))]

    for index, folder in enumerate(folders):
        folder_path = os.path.join(root_path, folder)

        if index == len(folders) - 1:
            print(indent + "└── " + folder)
            new_indent = indent + "    "
        else:
            print(indent + "├── " + folder)
            new_indent = indent + "│   "

        print_folder_structure(folder_path, new_indent)


# 👇 Use this path in your notebook
root_directory = r"../datasets/road"

print(root_directory)
print_folder_structure(root_directory)
```

```python
import os
from pathlib import Path

DATASETS = Path("../datasets").resolve()
IGNORE_FOLDERS = [".git", ".ipynb_checkpoints", "__MACOSX"]

IMG_EXT = [".jpg", ".jpeg", ".png", ".bmp"]
VID_EXT = [".mp4", ".avi", ".mov"]
YOLO_LABEL_EXT = ".txt"
VOC_LABEL_EXT = ".xml"

def ignore_path(p):
    return any(x in str(p) for x in IGNORE_FOLDERS)

def detect_dataset_type(dataset_path):
    """
    Detects the type of dataset: YOLO, VOC, Classification, Video, or Mixed
    """
    has_images = False
    has_videos = False
    has_yolo_labels = False
    has_voc_labels = False
    
    for base, dirs, files in os.walk(dataset_path, topdown=True):
        dirs[:] = [d for d in dirs if d not in IGNORE_FOLDERS]
        
        for f in files:
            f_lower = f.lower()
            if any(f_lower.endswith(ext) for ext in IMG_EXT):
                has_images = True
            elif any(f_lower.endswith(ext) for ext in VID_EXT):
                has_videos = True
            elif f_lower.endswith(YOLO_LABEL_EXT) and "labels" in base.lower():
                has_yolo_labels = True
            elif f_lower.endswith(VOC_LABEL_EXT):
                has_voc_labels = True
    
    # Determine type
    if has_images and has_yolo_labels:
        return "YOLO Detection"
    elif has_images and has_voc_labels:
        return "VOC Detection"
    elif has_images and not (has_yolo_labels or has_voc_labels):
        return "Classification"
    elif has_videos:
        return "Video"
    elif has_images and has_yolo_labels and has_voc_labels:
        return "Mixed Detection"
    else:
        return "Unknown / Other"

# Scan all datasets
dataset_types = {}
for dataset in DATASETS.iterdir():
    if dataset.is_dir():
        dtype = detect_dataset_type(dataset)
        dataset_types[dataset.name] = dtype

# Print results
print("\nDATASET TYPES:")
for ds, dtype in dataset_types.items():
    print(f"{ds} → {dtype}")
```

```python
import os
from pathlib import Path
from collections import Counter
import pandas as pd

# -----------------------------
# CONFIG
# -----------------------------
DATASETS_DIR = Path("../datasets").resolve()
IGNORE_FOLDERS = [".git", ".ipynb_checkpoints", "__MACOSX"]

IMG_EXT = [".jpg", ".jpeg", ".png", ".bmp"]
VID_EXT = [".mp4", ".avi", ".mov"]
YOLO_LABEL_EXT = ".txt"
VOC_LABEL_EXT = ".xml"

# -----------------------------
# HELPER FUNCTIONS
# -----------------------------
def ignore_path(p):
    return any(x in str(p) for x in IGNORE_FOLDERS)

def detect_dataset_type(dataset_path):
    """Detect dataset type based on contents"""
    has_images = False
    has_videos = False
    has_yolo_labels = False
    has_voc_labels = False
    
    for base, dirs, files in os.walk(dataset_path, topdown=True):
        dirs[:] = [d for d in dirs if d not in IGNORE_FOLDERS]
        for f in files:
            f_lower = f.lower()
            if any(f_lower.endswith(ext) for ext in IMG_EXT):
                has_images = True
            elif any(f_lower.endswith(ext) for ext in VID_EXT):
                has_videos = True
            elif f_lower.endswith(YOLO_LABEL_EXT) and "labels" in base.lower():
                has_yolo_labels = True
            elif f_lower.endswith(VOC_LABEL_EXT):
                has_voc_labels = True
    
    if has_images and has_yolo_labels:
        return "YOLO Detection"
    elif has_images and has_voc_labels:
        return "VOC Detection"
    elif has_images and not (has_yolo_labels or has_voc_labels):
        return "Classification"
    elif has_videos:
        return "Video"
    else:
        return "Unknown / Other"

def count_dataset_files(dataset_path):
    """Count images, videos, labels and YOLO classes"""
    image_count = 0
    video_count = 0
    label_count = 0
    class_counter = Counter()
    
    for base, dirs, files in os.walk(dataset_path, topdown=True):
        dirs[:] = [d for d in dirs if d not in IGNORE_FOLDERS]
        for f in files:
            f_lower = f.lower()
            f_path = Path(base) / f
            try:
                if any(f_lower.endswith(ext) for ext in IMG_EXT):
                    image_count += 1
                elif any(f_lower.endswith(ext) for ext in VID_EXT):
                    video_count += 1
                elif f_lower.endswith(YOLO_LABEL_EXT) and "labels" in base.lower():
                    label_count += 1
                    # count classes in YOLO file
                    try:
                        lines = f_path.read_text().splitlines()
                        for line in lines:
                            if len(line.strip()) == 0:
                                continue
                            cls = line.split()[0]
                            class_counter[cls] += 1
                    except:
                        continue
            except:
                continue
    
    return image_count, video_count, label_count, dict(class_counter)

# -----------------------------
# SCAN ALL DATASETS
# -----------------------------
dataset_summary = []

for dataset in sorted(DATASETS_DIR.iterdir(), key=lambda x: x.name.lower()):
    if not dataset.is_dir():
        continue
    dtype = detect_dataset_type(dataset)
    images, videos, labels, classes = count_dataset_files(dataset)
    dataset_summary.append({
        "Dataset": dataset.name,
        "Type": dtype,
        "Images": images,
        "Videos": videos,
        "YOLO Labels": labels,
        "Num Classes": len(classes),
        "Classes": classes
    })

# -----------------------------
# CREATE TABLES
# -----------------------------
df_summary = pd.DataFrame(dataset_summary)
pd.set_option("display.max_columns", None)
pd.set_option("display.max_colwidth", None)

print("\n========== DATASET SUMMARY ==========")
display(df_summary)

# -----------------------------
# Optional: Flatten class counts per dataset
# -----------------------------
rows = []
for row in dataset_summary:
    ds = row["Dataset"]
    for cls, cnt in row["Classes"].items():
        rows.append([ds, cls, cnt])

df_classes = pd.DataFrame(rows, columns=["Dataset", "Class", "Count"])
print("\n========== CLASS DISTRIBUTION ==========")
display(df_classes)
```
